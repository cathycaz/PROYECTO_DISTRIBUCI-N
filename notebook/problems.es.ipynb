# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

````python
from scipy.stats import binom

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
````

````python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)
print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
````

````python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
````

DISTRIBUCIÓN NORMAL

````python
from scipy.stats import norm

def dnorm(x, mean = 0, std = 1):
    """
    Calculates the density of the normal distribution
    """
    result = norm.pdf(x, loc = mean, scale = std)

    return result

def pnorm(x, mean = 0, std = 1):
    """
    Calculates the cumulative of the normal distribution
    """
    result = norm.cdf(x = x, loc = mean, scale = std)

    return result

def qnorm(p, mean = 0, std = 1):
    """
    Calculates the quantile function from the normal distribution
    """
    result = norm.ppf(q = p, loc = mean, scale = std)

    return result

def rnorm(n, mean = 0, std = 1):
    """
    Generates random variables from the normal distribution
    """
    result = norm.rvs(size = n, loc = mean, scale = std)

    return result
````

````python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(50)

mean = 0
std = 1
data = np.arange(-4, 4, 0.05)

pmf = dnorm(data, mean = mean, std = std)
cdf = pnorm(data, mean = mean, std = std)
ppf = qnorm(data, mean = mean, std = std)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].plot(data, pmf, "r-")
axis[1].plot(data, cdf, "b-")
axis[2].plot(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
````

DISTRIBUCIÓN UNIFORME

````python
from scipy.stats import uniform

def dunif(x, low = 0, high = 1):
    """
    Calculates the point estimate of the uniform distribution
    """
    result = uniform.pdf(x, loc = low, scale = (high - low))

    return result

def punif(q, low = 0, high = 1):
    """
    Calculates the cumulative of the uniform distribution
    """
    result = uniform.cdf(q, loc = low, scale = (high - low))

    return result

def qunif(p, low = 0, high = 1):
    """
    Calculates the quantile function from the uniform distribution
    """
    result = uniform.ppf(p, loc = low, scale = (high - low))

    return result

def runif(n, low = 0, high = 1):
    """
    Generates random variables from the uniform distribution
    """
    result = uniform.rvs(loc = low, scale = (high - low), size = n)

    return result
````

````python
np.random.seed(50)

low = 0
high = 1
data = np.arange(-5, 5, 0.01)

pmf = dunif(data, low = low, high = high)
cdf = punif(data, low = low, high = high)
ppf = qunif(data, low = low, high = high)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].plot(data, pmf, "r-")
axis[1].plot(data, cdf, "b-")
axis[2].plot(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
````

OBJETIVOS PEDAGÓGICOS DE ESTOS EJERCICIOS

COMPRENSIÓN DE LAS 4 FUNCIONES FUNDAMENTALES

Cada distribución tiene 4 funciones básicas que todos los estadísticos deben dominar:


Las 4 Funciones Universales:

d___() (PDF/PMF): "¿Qué tan probable es este valor?"

p___() (CDF): "¿Qué probabilidad hay hasta este punto?"

q___() (PPF): "¿Qué valor corresponde a este percentil?"

r___() (RVS): "Genera datos aleatorios"


LECCIÓN: Estas 4 funciones son el "vocabulario básico" de cualquier distribución estadística.





### **✅ SEÑALES PARA USAR BINOMIAL:**
- **Experimento:** Tiene solo 2 resultados posibles (éxito/fracaso)
- **Repetición:** Haces el mismo experimento n veces
- **Independencia:** Cada experimento no afecta a los otros
- **Probabilidad constante:** La probabilidad de éxito es la misma siempre

### **🎯 EJEMPLOS TÍPICOS:**
```
✓ Lanzar 10 monedas → ¿Cuántas caras?
✓ 100 pacientes toman medicina → ¿Cuántos se curan?
✓ 20 preguntas de verdadero/falso → ¿Cuántas correctas?
✓ 50 productos → ¿Cuántos defectuosos?
✓ 1000 emails → ¿Cuántos son spam?
```

### **❌ NO USES BINOMIAL SI:**
- Los resultados no son solo 2 opciones
- La probabilidad cambia en cada intento
- Los eventos están relacionados entre sí

---

## 🔔 DISTRIBUCIÓN NORMAL - Úsala cuando:

### **✅ SEÑALES PARA USAR NORMAL:**
- **Datos naturales:** Resultado de muchos factores aleatorios
- **Simetría:** Los datos se concentran alrededor de un valor central
- **Teorema Central:** Suma/promedio de muchas variables
- **Errores:** Mediciones con error aleatorio

### **🎯 EJEMPLOS TÍPICOS:**
```
✓ Altura de personas → Normal(170cm, 10cm)
✓ Puntuaciones de IQ → Normal(100, 15)
✓ Errores de medición → Normal(0, σ)
✓ Notas de examen (bien diseñado) → Normal(μ, σ)
✓ Presión arterial → Normal(120, 15)
✓ Tiempo de reacción → Normal(μ, σ)
```

### **❌ NO USES NORMAL SI:**
- Los datos están limitados (no pueden ser negativos)
- Los datos son muy asimétricos
- Los datos tienen múltiples picos

---

## 📏 DISTRIBUCIÓN UNIFORME - Úsala cuando:

### **✅ SEÑALES PARA USAR UNIFORME:**
- **Azar puro:** Todos los valores son igualmente probables
- **Rango fijo:** Los valores están entre un mínimo y máximo
- **Sin preferencia:** No hay valores más probables que otros

### **🎯 EJEMPLOS TÍPICOS:**
```
✓ Lanzar un dado equilibrado → Uniforme(1, 6)
✓ Hora de llegada del autobús → Uniforme(9:00, 9:30)
✓ Elegir número al azar → Uniforme(0, 100)
✓ Generador de números aleatorios → Uniforme(0, 1)
✓ Posición inicial en una carrera → Uniforme(1, n)
```

### **❌ NO USES UNIFORME SI:**
- Algunos valores son más probables que otros
- Los datos muestran tendencia central
- Hay límites naturales en los datos

---

## 🧠 PROCESO DE DECISIÓN PASO A PASO

### **1. PREGÚNTATE:**
- ¿Estoy contando éxitos/fracasos? → **BINOMIAL**
- ¿Es una medición natural con variabilidad? → **NORMAL**
Que ocurre en la naturaleza o en la vida real, Que no todos tienen el mismo valor
- ¿Todos los valores son igualmente probables? → **UNIFORME**

### **2. VERIFICA LOS SUPUESTOS:**
- **Binomial:** n fijo, p constante, independencia
- **Normal:** simetría, datos continuos, teorema central límite
- **Uniforme:** rango fijo, probabilidad constante

### **3. ANALIZA LOS DATOS:**
- **Histograma:** ¿Qué forma tienen?
- **Rango:** ¿Cuáles son los límites?
- **Contexto:** ¿De dónde vienen los datos?

**RESPUESTA:** UNIFORME DISCRETA(1, 6)

## ⚠️ ERRORES COMUNES A EVITAR

### **❌ ERROR 1: Usar Normal para conteos pequeños**
- **Problema:** "Número de accidentes por día" → Normal
- **Correcto:** Usar Poisson o Binomial

### **❌ ERROR 2: Usar Binomial sin verificar independencia**
- **Problema:** "Estudiantes que aprueban en la misma clase" → Binomial
- **Correcto:** Puede haber dependencia (mismo profesor, mismo material)

### **❌ ERROR 3: Usar Uniforme cuando hay tendencia central**
- **Problema:** "Calificaciones de estudiantes" → Uniforme
- **Correcto:** Normalmente hay más estudiantes en el rango medio

### **❌ ERROR 4: Usar Normal para datos con límites estrictos**
- **Problema:** "Porcentajes (0-100%)" → Normal
- **Correcto:** Beta o Uniforme, ya que Normal puede dar valores negativos
